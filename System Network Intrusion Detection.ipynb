### Multinomial Classification Model to detect type of attack

```python
# We will use labelencoding to categorize based on different kinds of attacks

from sklearn.preprocessing import LabelEncoder
```

```python
final.attack.unique()
```

```python
cols = ['attack']
mapping_dict = {}
for i in cols:
    le = LabelEncoder()
    final[i] = le.fit_transform(final[i])
    d = dict(zip(le.classes_,list(range(0,len(le.classes_)))))  # here we will create a dict to store the encoding values
    mapping_dict.update(d)
```
